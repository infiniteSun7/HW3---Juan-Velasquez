1) Write program with while loop which finds the followings for the given values between 150 to 1000.
a) How many odd numbers are between 150 and 1000 (include 150 and 1000)
b) Find the sum of odd numbers which are between 150 and 1000 (include 150 and 1000)
c) Find the multiplication of odd numbers which are between 150 and 1000 (include 150
and 1000)
d) Find the sum of square of odd numbers which are between 150 and 1000 (include 150
and 1000)


```python
num = 150
count = 0
total_sum = 0
product = 1
sum_squares = 0

while num <= 1000:
    if num % 2 != 0:  # Check if number is odd
        count += 1
        total_sum += num
        product *= num
        sum_squares += num**2
    num += 1

print(f"Count of odd numbers: {count}")
print(f"Sum of odd numbers: {total_sum}")
print(f"Multiplication (Product): {product}")
print(f"Sum of squares: {sum_squares}")
```

2) Write program with for loop which finds the followings for the given values between 150 to 1000.
a) How many odd numbers are between 150 and 1000 (include 150 and 1000)
b) Find the sum of odd numbers which are between 150 and 1000 (include 150 and 1000)
c) Find the multiplication of odd numbers which are between 150 and 1000 (include 150
and 1000)
d) Find the sum of square of odd numbers which are between 150 and 1000 (include 150
and 1000)


```python
count = 0
total_sum = 0
product = 1
sum_squares = 0

# range(start, stop) - stop is exclusive, so we use 1001
for num in range(150, 1001):
    if num % 2 != 0:
        count += 1
        total_sum += num
        product *= num
        sum_squares += num**2

print(f"Count: {count}")
print(f"Sum: {total_sum}")
print(f"Product: {product}")
print(f"Sum of squares: {sum_squares}")
```

3) Write a program so that for given M, it computes the sum . 1 1 M k s k = = 
(Hint: Use the while-loop . As we see in in the class, you can update s as s=s+k and you need to increase the k in each step such that k=k+1).


```python
M = int(input("Enter value for M: "))
s = 5
k = 5

while k <= M:
    s = s + k
    k = k + 1

print(f"The sum from 1 to {M} is: {s}")
```

4) Write a program which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number. For example:
Enter a number: 4
Enter a number: 5
Enter a number: bad data Invalid input
Enter a number: 7
Enter a number: done
16 3 5.333333333333333


```python
total = 0
count = 0

while True:
    val = input("Enter a number: ")
    if val.lower() == "done":
        break
    try:
        f_val = float(val)
        total += f_val
        count += 1
    except ValueError:
        print("Invalid input")
        continue

if count > 0:
    print(total, count, total / count)
else:
    print("No numbers were entered.")
```

5) Write another program that prompts for a list of numbers as above (question 4) and at the end prints out both the maximum and minimum of the numbers instead of the average.

```python
largest = None
smallest = None

while True:
    val = input("Enter a number: ")
    if val.lower() == "done":
        break
    try:
        num = float(val)
        if largest is None or num > largest:
            largest = num
        if smallest is None or num < smallest:
            smallest = num
    except ValueError:
        print("Invalid input")

print(f"Maximum: {largest}")
print(f"Minimum: {smallest}")
```
